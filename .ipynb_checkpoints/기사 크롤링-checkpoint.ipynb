# 월(月) 크롤링

```python
def month():
    df_oo = []
    news_date = []
    news_title = []
    news_contents = []
    
    cnt = 0

    start_date = input('시작 날짜를 입력하세요')
    end_date = input('끝 날짜를 입력하세요')

    for p in name:
        try:
            for i in range(1, 6):

                url = f'https://search.naver.com/search.naver?where=news&sm=tab_pge&query={p}&sort=0&photo=0&field=0&pd=3&ds={start_date}&de={end_date}&cluster_rank=151&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:from20170101to20170131,a:all&start={(i - 1) * 10 + 1}'

                res = requests.get(url)
                soup = BeautifulSoup(res.text, 'html.parser')

                for j in range(10):

                    date = soup.select('span.info')[j].text
                    title = soup.select('a.news_tit')[j].text
                    contents = soup.select('div.news_dsc')[j].text
                    news_date.append(date)
                    news_title.append(title)
                    news_contents.append(contents)
#                     print(date)
#                     print(title)
#                     print(contents)
#                     print('='*50)
                    globals()['df_'+str(p)] = pd.DataFrame({'기업' : p, '날짜' : news_date, '제목' : news_title, '내용' : news_contents})

        except:
            globals()['df_'+str(p)] = pd.DataFrame({'기업' : p, '날짜' : news_date, '제목' : news_title, '내용' : news_contents})
        
        cnt += 1
        print(cnt)
        print('★★★★★★★★★★★★★', p, '★★★★★★★★★★★★★')
        df_oo.append(globals()['df_'+str(p)])        

    globals()['df_news_'+str(start_date.replace('.', ''))] = pd.concat(df_oo) 
```
