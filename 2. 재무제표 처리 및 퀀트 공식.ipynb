### 퀀트(Quant)

```python
df_financial = pd.read_excel('최종재무제표.xlsx')
df_financial = df_financial.drop('Unnamed: 0', axis = 1)
df_financial
```

```python
df_financial.info()
```

```python
df_financial.columns
```

```python
# 연간 수익률

yearly_price_df = df_financial.pivot_table(index='year', columns = 'name', values='수정주가')
yearly_price_df
```

```python
yearly_rtn_df = yearly_price_df.pct_change().shift(-1)
yearly_rtn_df
```

## 퀀트 공식 1
* PER : 0 ~ 10
* PBR : 0 ~ 1
* PCR : 0 ~ 5
* ROE : 10% 이상

```python
# PER : 0 ~ 10

df_financial = df_financial[(df_financial['PER(배)'] > 0) & (df_financial['PER(배)'] < 10)]
```

```python
# PBR : 0 ~ 1

df_financial = df_financial[(df_financial['PBR(배)'] > 0) & (df_financial['PBR(배)'] < 1)]
```

```python
# PCR : 0 ~ 5

df_financial = df_financial[(df_financial['PCR'] > 0) & (df_financial['PCR'] < 5)]
```

```python
# ROE : 10% 이상

df_financial = df_financial[df_financial['ROE(%)'] >= 0.1]
```

```python
df_financial
```

```python
df_financial['year'].unique()
```

```python
selected_df = df_financial.pivot(index='year', columns='name', values='PBR(배)')
selected_df
```

```python
asset_on_df = selected_df.notna().astype(int).replace(0, np.nan)
asset_on_df
```

```python
# 연도별 투자 수익률

rtn_series = (yearly_rtn_df * asset_on_df).mean(axis=1)
rtn_series
```

```python
# 연도별 누적 수익률

cum_rtn_str_df = (rtn_series + 1).cumprod()
cum_rtn_str_df
```

```python
cum_rtn_str_df.plot(figsize=(10,3))
```
